The Constants in math module are ```math.pi``` and ```math.e```

The ```math.exp()``` function in Python returns the value of e (approximately 2.71828) raised to the power of the given argument.

The ```math.pow()``` function in Python returns the value of the first argument raised to the power of the second argument.

````python
import math

number = 10.0
natural_log = math.log(number)
print("Natural logarithm of", number, "is:", natural_log)
````
