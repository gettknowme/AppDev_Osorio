**EXERCISE 3**

Create a list containing the valid grade values, then using enumerate in a for loop, print out all indices and items in the list.

It should print out:
```
0   1.0
1   1.25
2   1.5
3   1.75
4   2.0
5   2.25
6   2.5
7   3.0
```

````python
grades = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]

#  for loop to print out all indices and items in the list
for index, grade in enumerate(grades):
    print(f"{index}   {grade}")
````

**EXERCISE 4**

Using Index Slicing, print out the list of grades higher than 2.0

````python
grades = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 3.0]
print(grades[grades.index(2.0):])
````

**CHALLENGE**

Modify the code in the subheading Parsing Lines to display all the time stamps of the emails in mboxt-short.txt

````python
import re


with open('mboxt-short.txt', 'r') as file:

    timestamps = []


    for line in file:
        stamps
        match = re.search(r'^Date: .*?(\d{2} \w{3} \d{4} \d{2}:\d{2}:\d{2})', line)
        if match:
            # Extract the matched time stamp
            timestamp = match.group(1)
            # Append the time stamp to the list
            timestamps.append(timestamp)


for timestamp in timestamps:
    print(timestamp)
````
